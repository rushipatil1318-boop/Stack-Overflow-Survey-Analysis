```python
import urllib.request
urllib.request.urlretrieve('https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv', 'survey-data.csv')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("survey-data.csv")
```

```python
df.head()
```

```python
age_comp = df.groupby('Age')['ConvertedCompYearly'].median()
age_comp.plot(kind='line', figsize=(10, 6))
plt.title('Median Compensation by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Median Compensation')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
age_25_45 = df[df['Age'].isin(['25-34 years old', '35-44 years old'])]['ConvertedCompYearly'].groupby(df['Age']).median()
age_25_45.plot(kind='line', figsize=(10, 6))
plt.title('Median Compensation for Ages 25 to 45')
plt.xlabel('Age Group')
plt.ylabel('Median Compensation')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
exp_sat = df.groupby('WorkExp')['JobSatPoints_6'].mean()
exp_sat.plot(kind='line', figsize=(10, 6))
plt.title('Job Satisfaction by Experience Level')
plt.xlabel('Work Experience (Years)')
plt.ylabel('Average Job Satisfaction')
plt.grid(True)
plt.show()
```

```python
exp_comp = df.groupby('WorkExp')['ConvertedCompYearly'].median()
exp_comp.plot(kind='line', figsize=(10, 6))
plt.title('Median Compensation Over Experience Level')
plt.xlabel('Work Experience (Years)')
plt.ylabel('Median Compensation')
plt.grid(True)
plt.show()
```

```python
exp_sat2 = df.groupby('WorkExp')['JobSatPoints_6'].mean()
exp_sat2.plot(kind='line', figsize=(10, 6))
plt.title('Job Satisfaction Across Experience Levels')
plt.xlabel('Work Experience (Years)')
plt.ylabel('Average Job Satisfaction')
plt.grid(True)
plt.show()
```
