```python
# Load dataset directly from URL (Windows-compatible)
import pandas as pd
import matplotlib.pyplot as plt

file_path = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv"
df = pd.read_csv(file_path)

print("Dataset shape:", df.shape)
print("\nFirst few rows:")
df.head()
```

```python
# Task 1.1: Pie Chart for Top 5 Databases Respondents Want to Work With
# Extract databases from the column
databases = []
for db_list in df['DatabaseWantToWorkWith'].dropna():
    if pd.notna(db_list):
        databases.extend(str(db_list).split(';'))

# Count frequencies
from collections import Counter
db_counts = Counter([db.strip() for db in databases if db.strip()])

# Get top 5
top_5_dbs = dict(db_counts.most_common(5))

# Create pie chart
plt.figure(figsize=(10, 7))
colors = plt.cm.Set3(range(len(top_5_dbs)))
plt.pie(top_5_dbs.values(), labels=top_5_dbs.keys(), autopct='%1.1f%%', 
        colors=colors, startangle=90, textprops={'fontsize': 11})
plt.title('Top 5 Databases Respondents Want to Work With', fontsize=14, fontweight='bold', pad=20)
plt.tight_layout()
plt.show()
```

```python
# Task 1.2: Pie Chart for Top 5 Developer Roles
# Extract developer types
dev_types = []
for devs in df['DevType'].dropna():
    if pd.notna(devs):
        dev_types.extend(str(devs).split(';'))

# Count frequencies
from collections import Counter
dev_counts = Counter([dev.strip() for dev in dev_types if dev.strip()])

# Get top 5
top_5_devs = dict(dev_counts.most_common(5))

# Create pie chart
plt.figure(figsize=(10, 7))
colors = plt.cm.Pastel1(range(len(top_5_devs)))
plt.pie(top_5_devs.values(), labels=[d[:25] for d in top_5_devs.keys()], autopct='%1.1f%%',
        colors=colors, startangle=90, textprops={'fontsize': 10})
plt.title('Top 5 Developer Roles', fontsize=14, fontweight='bold', pad=20)
plt.tight_layout()
plt.show()
```

```python
# Task 1.3: Pie Chart for Operating Systems Used Professionally
# Extract operating systems
os_types = []
for oss in df['OpSysProfessional use'].dropna():
    if pd.notna(oss):
        os_types.extend(str(oss).split(';'))

# Count frequencies
from collections import Counter
os_counts = Counter([os.strip() for os in os_types if os.strip()])

# Get top 5
top_5_os = dict(os_counts.most_common(5))

# Create pie chart
plt.figure(figsize=(10, 7))
colors = plt.cm.Set2(range(len(top_5_os)))
plt.pie(top_5_os.values(), labels=top_5_os.keys(), autopct='%1.1f%%',
        colors=colors, startangle=45, textprops={'fontsize': 10})
plt.title('Top Operating Systems Used Professionally', fontsize=14, fontweight='bold', pad=20)
plt.tight_layout()
plt.show()
```

```python
 

languages = []
for lang_list in df['LanguageHaveWorkedWith'].dropna():
    if pd.notna(lang_list):
        languages.extend(str(lang_list).split(';'))

from collections import Counter
lang_counts = Counter([lang.strip() for lang in languages if lang.strip()])


top_5_langs = dict(lang_counts.most_common(5))


plt.figure(figsize=(10, 7))
colors = plt.cm.Set3(range(len(top_5_langs)))
explode = [0.05] * len(top_5_langs)
plt.pie(top_5_langs.values(), labels=top_5_langs.keys(), autopct='%1.1f%%',
        colors=colors, explode=explode, startangle=90, textprops={'fontsize': 11})
plt.title('Top 5 Programming Languages Respondents Have Worked With', fontsize=14, fontweight='bold', pad=20)
plt.tight_layout()
plt.show()
```

```python

ai_col = None
for col in df.columns:
    if 'AITool' in col or 'ai_tool' in col.lower():
        if 'current' in col.lower():
            ai_col = col
            break

if ai_col is None:
    ai_col = [col for col in df.columns if 'AITool' in col][0] if any('AITool' in col for col in df.columns) else None

if ai_col:
    ai_tools = []
    for tool_list in df[ai_col].dropna():
        if pd.notna(tool_list):
            ai_tools.extend(str(tool_list).split(';'))
    
    from collections import Counter
    ai_counts = Counter([tool.strip() for tool in ai_tools if tool.strip()])
    
    top_5_ai = dict(ai_counts.most_common(5))
    
    if top_5_ai:
        
        plt.figure(figsize=(10, 7))
        colors = plt.cm.OrRd(range(len(top_5_ai)))
        plt.pie(top_5_ai.values(), labels=[t[:20] for t in top_5_ai.keys()], autopct='%1.1f%%',
                colors=colors, startangle=45, textprops={'fontsize': 10})
        plt.title('Top 5 AI Development Tools Currently Using', fontsize=14, fontweight='bold', pad=20)
        plt.tight_layout()
        plt.show()
    else:
        print("No AI tools data available")
else:
    print("AI tools column not found")
```
